## Built-in Functions

```
print('Asabeneh', 'Yetayeh', 'Finland', sep='  ')
print(len('Python'))
print(type(10))
print(str(10))
print(int('10'))
print(float('9.8'))
print(min(2, 3, 4, 5))
print(max(2, 3, 4, 5))
print(abs(-10))
print(sum([1, 2, 3, 4, 5]))
```

## Variables

```
first_name = 'Asabeneh'
last_name = 'Yetayeh'
country = 'Finland'
city = 'Helsinki'
age = 250
is_married = True
skills = ['HTML', 'CSS', 'JS', 'React', 'Python']
person_info = {
    'firstname': 'Asabeneh',
    'lastname': 'Yetayeh',
    'country': 'Finland',
    'city': 'Helsinki'
}
print('First name:', first_name)
print('Last name:', last_name)
print('Country:', country)
print('City:', city)
print('Age:', age)
print('Married:', is_married)
print('Skills:', skills)
print('Person info:', person_info)
```

## Declaring Multiple Variables in a Line

```
first_name, last_name, country, age, is_married = 'Asabeneh', 'Yetayeh', 'Finland', 250, True
print(first_name, last_name, country, age, is_married)
```
