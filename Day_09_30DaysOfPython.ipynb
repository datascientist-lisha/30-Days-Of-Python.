## if Condition

```
a = 3
if a > 0:
    print('A is a positive number')
print('Outside the scope')
```

## if else

```
a = 3
if a < 0:
    print('A is a negative number')
else:
    print('A is a positive number')
```

## if elif else

```
a = 0
if a > 0:
    print('A is a positive number')
elif a < 0:
    print('A is a negative number')
else:
    print('A is zero')
```

## Short Hand

```
a = 3
print('A is positive') if a > 0 else print('A is negative')
```

## Exercises: Level 1

```
# 1. Get user input using input("Enter your age: ").
# If age >= 18, print 'You are old enough to drive'
# If below 18, print something else
age = int(input('Enter your age: '))
if age >= 18:
    print('You are old enough to drive')
else:
    print('You are not old enough to drive')
```

```
# 2. Compare the values of my_age and your_age using if else
my_age = 250
your_age = int(input('Enter your age: '))
if my_age > your_age:
    print(f'I am {my_age - your_age} years older than you')
elif my_age < your_age:
    print(f'I am {your_age - my_age} years younger than you')
else:
    print('We are the same age')
```

```
# 3. Get two numbers from the user
# If a is greater than b return a is greater than b
# If a is less than b return a is smaller than b
# If they are equal return a is equal to b
a = float(input('Enter a: '))
b = float(input('Enter b: '))
if a > b:
    print('a is greater than b')
elif a < b:
    print('a is smaller than b')
else:
    print('a is equal to b')
```

## Exercises: Level 2

```
# 1. Write a code which gives grade to students by their score
score = float(input('Enter score: '))
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f'Grade: {grade}')
```

```
# 2. Check if the season is Autumn, Winter, Spring or Summer
month = int(input('Enter month (1-12): '))
if month in [9, 10, 11]:
    print('Autumn')
elif month in [12, 1, 2]:
    print('Winter')
elif month in [3, 4, 5]:
    print('Spring')
else:
    print('Summer')
```
