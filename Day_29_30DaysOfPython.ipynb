## Flask REST API

```
api_code = '''
from flask import Flask, jsonify, request
app = Flask(__name__)

students = [
    {'id': 1, 'name': 'Asabeneh', 'country': 'Finland', 'city': 'Helsinki', 'skills': ['HTML', 'CSS', 'Python']},
    {'id': 2, 'name': 'David', 'country': 'UK', 'city': 'London', 'skills': ['Python', 'MongoDB']}
]

@app.route('/api/v1.0/students', methods=['GET'])
def get_students():
    return jsonify(students)

@app.route('/api/v1.0/students/<int:id>', methods=['GET'])
def get_student(id):
    student = [s for s in students if s['id'] == id]
    return jsonify(student[0]) if student else jsonify({'message': 'Not found'}), 404

if __name__ == '__main__':
    app.run(debug=True)
'''
print(api_code)
```
