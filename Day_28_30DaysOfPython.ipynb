## Requesting APIs

```
import requests
url = 'https://restcountries.com/v3.1/all'
try:
    response = requests.get(url)
    countries = response.json()
    print('Number of countries:', len(countries))
    print('First country:', countries[0]['name']['common'])
except Exception as e:
    print(e)
```

## Exercises: Level 1

```
import requests
url = 'https://api.github.com/users/Asabeneh'
try:
    r = requests.get(url)
    data = r.json()
    print('Name:', data.get('name'))
    print('Public repos:', data.get('public_repos'))
except Exception as e:
    print(e)
```

## Exercises: Level 2

```
url = 'https://restcountries.com/v3.1/name/finland'
try:
    r = requests.get(url)
    data = r.json()[0]
    print('Country:', data['name']['common'])
    print('Capital:', data['capital'][0])
    print('Population:', data['population'])
except Exception as e:
    print(e)
```
