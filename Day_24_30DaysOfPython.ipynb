## Statistics Module

```
import statistics
ages = [20, 20, 4, 24, 25, 22, 26, 20, 23, 22, 26]
print('Mean:', statistics.mean(ages))
print('Median:', statistics.median(ages))
print('Mode:', statistics.mode(ages))
print('Stdev:', statistics.stdev(ages))
print('Variance:', statistics.variance(ages))
```

## Exercises: Level 1

```
# Use the statistics module
data = [89, 90, 85, 90, 88, 95, 72, 65, 55, 100]
print('Mean:', statistics.mean(data))
print('Median:', statistics.median(data))
print('Mode:', statistics.mode(data))
print('Stdev:', statistics.stdev(data))
print('Variance:', statistics.variance(data))
```

## Exercises: Level 2

```
import random
random.seed(0)
ages = [random.randint(1, 100) for _ in range(100)]
print('Mean:', statistics.mean(ages))
print('Median:', statistics.median(ages))
print('Min:', min(ages))
print('Max:', max(ages))
```

```
# 1. Convert the ages into frequency table
from collections import Counter
ages = [20, 20, 4, 24, 25, 22, 26, 20, 23, 22, 26]
freq = Counter(ages)
print('Frequency table:', dict(sorted(freq.items())))
```
