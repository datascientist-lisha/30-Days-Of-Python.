## while Loop

```
count = 0
while count < 5:
    print(count)
    count = count + 1
```

## break and continue

```
count = 0
while count < 5:
    if count == 3:
        break
    print(count)
    count = count + 1
```

```
count = 0
while count < 5:
    if count == 3:
        count = count + 1
        continue
    print(count)
    count = count + 1
```

## for Loop

```
for number in range(11):
    print(number)
```

```
numbers = [0, 1, 2, 3, 4, 5]
for number in numbers:
    print(number)
```

```
for i in range(0, 11, 2):
    print(i)
```

## Exercises: Level 1

```
# 1. Iterate 0 to 10 using for loop, do the same using while loop
for i in range(11):
    print(i)
```

```
i = 0
while i <= 10:
    print(i)
    i += 1
```

```
# 2. Iterate 10 to 0 using for loop, do the same using while loop
for i in range(10, -1, -1):
    print(i)
```

```
# 3. Write a loop that makes seven calls to print(), printing
# the following triangle
for i in range(1, 8):
    print('#' * i)
```

```
# 4. Use nested loops to create a multiplication table
for i in range(1, 8):
    for j in range(1, 8):
        print(f'{i} x {j} = {i*j}', end='	')
    print()
```

## Exercises: Level 2

```
# 1. Use for loop to iterate from 0 to 100 and print only even numbers
for i in range(0, 101):
    if i % 2 == 0:
        print(i)
```

```
# 2. Use for loop to iterate from 0 to 100 and print only odd numbers
for i in range(0, 101):
    if i % 2 != 0:
        print(i)
```

```
# 3. Use for loop to iterate from 0 to 100 and print the sum of all numbers
total = 0
for i in range(101):
    total += i
print('Sum:', total)
```

```
# 4. Use for loop to iterate from 0 to 100 and print the sum of evens and odds
sum_evens = 0
sum_odds = 0
for i in range(101):
    if i % 2 == 0:
        sum_evens += i
    else:
        sum_odds += i
print('Sum of evens:', sum_evens)
print('Sum of odds:', sum_odds)
```

## Exercises: Level 3

```
# 1. Go through the countries list and extract all countries containing the word 'land'
countries = ['Finland', 'Iceland', 'Ireland', 'Germany', 'Netherlands', 'Poland']
land_countries = [c for c in countries if 'land' in c]
print(land_countries)
```

```
# 2. This is a fruit list, loop through the list and count the number of o's in each fruit
fruits = ['banana', 'orange', 'mango', 'lemon']
for fruit in fruits:
    print(f'{fruit}: {fruit.count("o")} o's')
```
