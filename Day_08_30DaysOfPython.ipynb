## Creating Dictionaries

```
empty_dict = {}
print(empty_dict)
dct = {'key1': 'value1', 'key2': 'value2', 'key3': 'value3', 'key4': 'value4'}
print(dct)
```

```
person = {
    'first_name': 'Asabeneh',
    'last_name': 'Yetayeh',
    'age': 250,
    'country': 'Finland',
    'is_married': True,
    'skills': ['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address': {
        'street': 'Space street',
        'zipcode': '02210'
    }
}
print(person)
```

## Accessing Dictionary Items

```
print(person['first_name'])
print(person['country'])
print(person['skills'])
print(person['skills'][0])
print(person['address']['street'])
print(person.get('first_name'))
print(person.get('country'))
print(person.get('salary'))
```

## Dictionary Methods

```
print(person.keys())
print(person.values())
print(person.items())
person.update({'job_title': 'Instructor', 'skills': ['HTML', 'CSS', 'JS', 'React', 'Python']})
print(person)
```

```
# 1. Write a function that checks if a key exists in a dictionary
def check_key(d, key):
    if key in d:
        return True
    return False

print(check_key(student, 'first_name'))
print(check_key(student, 'salary'))
```
