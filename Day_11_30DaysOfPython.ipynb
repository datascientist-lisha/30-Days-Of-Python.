## Defining a Function

```
def generate_full_name():
    first_name = 'Asabeneh'
    last_name = 'Yetayeh'
    space = ' '
    full_name = first_name + space + last_name
    print(full_name)

generate_full_name()
```

## Function with Parameters

```
def generate_full_name(first_name, last_name):
    space = ' '
    full_name = first_name + space + last_name
    return full_name

print(generate_full_name('Asabeneh', 'Yetayeh'))
print(generate_full_name('David', 'Smith'))
```

## Default Parameters

```
def greet_user(firstname, lastname='Yetayeh'):
    space = ' '
    full_name = firstname + space + lastname
    return full_name

print(greet_user('Asabeneh'))
print(greet_user('David', 'Smith'))
```

## *args and **kwargs

```
def sum_all_nums(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_all_nums(2, 3, 5))
print(sum_all_nums(2, 3, 5, 7, 9))
```

```
def print_person(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} = {value}')

print_person(name='Asabeneh', age=250, country='Finland')
```

## Lambda Function

```
square = lambda x: x ** 2
cube = lambda x: x ** 3
add_two_nums = lambda x, y: x + y

print(square(3))
print(cube(3))
print(add_two_nums(2, 3))
```

```
# 2. Area of a circle
def area_of_circle(r):
    import math
    return math.pi * r ** 2

print(area_of_circle(10))
```

```
# 3. Write a function add_all_nums which takes arbitrary number of args
def add_all_nums(*args):
    return sum(args)

print(add_all_nums(1, 2, 3, 4, 5))
```

```
# 4. Temperature converter
def convert_celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def convert_fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

print(convert_celsius_to_fahrenheit(0))
print(convert_fahrenheit_to_celsius(32))
```

## Exercises: Level 2

```
# 1. Declare a function called check_season
def check_season(month):
    if month in ['September', 'October', 'November']:
        return 'Autumn'
    elif month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    else:
        return 'Summer'

print(check_season('January'))
print(check_season('June'))
```

```
# 2. Declare a function called calculate_slope
def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

print(calculate_slope(2, 2, 6, 10))
```

## Exercises: Level 3

```
# 1. Declare a function called is_prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

for i in range(2, 20):
    if is_prime(i):
        print(i)
```

```
# 2. Declare functions which can do mathematical operations
def evens_and_odds(n):
    evens = [i for i in range(n+1) if i % 2 == 0]
    odds = [i for i in range(n+1) if i % 2 != 0]
    print('Evens:', evens)
    print('Odds:', odds)

evens_and_odds(100)
```
