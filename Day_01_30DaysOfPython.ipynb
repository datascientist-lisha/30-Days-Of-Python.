## Data Types

```
print(2 + 3)   # addition
print(3 - 1)   # subtraction
print(2 * 3)   # multiplication
print(3 / 2)   # division
print(3 ** 2)  # exponential
print(3 % 2)   # modulus
print(3 // 2)  # floor division
```

```
print(type(10))
print(type(3.14))
print(type(1 + 3j))
print(type('Asabeneh'))
print(type([1, 2, 3]))
print(type({'name': 'Asabeneh'}))
print(type({9.8, 3.14, 2.7}))
print(type((9.8, 3.14, 2.7)))
```
