## statistics Module

```
from statistics import mean, median, mode, stdev, variance
ages = [20, 20, 4, 24, 25, 22, 26, 20, 23, 22, 26]
print('Mean:', mean(ages))
print('Median:', median(ages))
print('Mode:', mode(ages))
print('Stdev:', stdev(ages))
print('Variance:', variance(ages))
```

## Exercises: Level 1

```
# 1. Writ a function which generates a six digit/character random_user_id
import random
import string

def random_user_id():
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(6))

print(random_user_id())
print(random_user_id())
print(random_user_id())
```

```
# 3. Write a function called rgb_color_gen which generates rgb colors
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f'rgb({r}, {g}, {b})'

print(rgb_color_gen())
```

## Exercises: Level 2

```
# 1. Write a function list_of_hexa_colors which returns any number of hexadecimal colors
def list_of_hexa_colors(num):
    colors = []
    for _ in range(num):
        color = '#' + ''.join([random.choice('0123456789ABCDEF') for _ in range(6)])
        colors.append(color)
    return colors

print(list_of_hexa_colors(3))
```

```
# 2. Write a function list_of_rgb_colors which returns any number of RGB colors
def list_of_rgb_colors(num):
    return [rgb_color_gen() for _ in range(num)]

print(list_of_rgb_colors(3))
```
