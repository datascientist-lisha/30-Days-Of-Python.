## SyntaxError

```
# SyntaxError example (commented to avoid stopping the notebook)
# print 'hello'  # Missing parentheses
print('SyntaxError: invalid syntax if you use print without parentheses')
```

## NameError

```
# NameError example
try:
    print(undeclared_variable)
except NameError as e:
    print(e)
```

## IndexError

```
try:
    numbers = [1, 2, 3]
    print(numbers[5])
except IndexError as e:
    print(e)
```

## ModuleNotFoundError

```
try:
    import maths
except ModuleNotFoundError as e:
    print(e)
```

## AttributeError

```
try:
    import math
    print(math.PI)
except AttributeError as e:
    print(e)
```

## KeyError

```
try:
    user = {'name': 'Asabeneh', 'age': 250}
    print(user['salary'])
except KeyError as e:
    print(e)
```

## TypeError

```
try:
    print(4 + '3')
except TypeError as e:
    print(e)
```

## ValueError

```
try:
    int('12a')
except ValueError as e:
    print(e)
```

## ZeroDivisionError

```
try:
    print(1/0)
except ZeroDivisionError as e:
    print(e)
```

## Exercises: Level 1

```
# Open Python interactive shell and try all the error types
errors = {
    'SyntaxError': 'raised when there is a syntax error',
    'NameError': 'raised when a name is not found',
    'IndexError': 'raised when the index is out of range',
    'ModuleNotFoundError': 'raised when a module is not found',
    'AttributeError': 'raised when an attribute is not found',
    'KeyError': 'raised when a key is not found in a dictionary',
    'TypeError': 'raised when there is a type mismatch',
    'ImportError': 'raised when an import fails',
    'FileNotFoundError': 'raised when a file is not found',
    'ValueError': 'raised when there is a wrong value',
    'ZeroDivisionError': 'raised when dividing by zero'
}
for error, description in errors.items():
    print(f'{error}: {description}')
```
