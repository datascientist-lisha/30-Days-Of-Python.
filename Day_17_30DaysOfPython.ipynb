## try and except

```
try:
    code_in_try_block
except TypeError as e:
    print(e)
```

```
try:
    name = input('Enter your name: ')
    year_born = input('Year you were born: ')
    age = 2020 - int(year_born)
    print(f'You are {name}. And your age is {age}.')
except TypeError:
    print('Type error occur')
except ValueError:
    print('Value error occur')
except ZeroDivisionError:
    print('zero division error occur')
```

## Packing and Unpacking with Exception

```
try:
    name = input('Enter your name: ')
    year_born = input('Year you were born: ')
    age = 2020 - int(year_born)
    print(f'You are {name}. And your age is {age}.')
except Exception as e:
    print(e)
```

## try, except, else and finally

```
try:
    name = input('Enter your name: ')
    year_born = input('Year you were born: ')
    age = 2020 - int(year_born)
    print(f'You are {name}. And your age is {age}.')
except TypeError as te:
    print('Type error occur')
    print(te)
except ValueError as ve:
    print('Value error occur')
    print(ve)
except ZeroDivisionError as ze:
    print('zero division error occur')
    print(ze)
else:
    print('I usually run with the try block')
finally:
    print('I always run.')
```

## Exercises: Level 1

```
# 1. Unpack the following tuple into four variables
country, year, city, population = ('Finland', 2020, 'Helsinki', 3500000)
print(country, year, city, population)
```

```
# 2. Unpack the engineers and the chief from staff
staff = [('Alice', 'Engineer'), ('Bob', 'Engineer'), ('Carol', 'Chief')]
*engineers, chief = staff
print('Engineers:', engineers)
print('Chief:', chief)
```

```
# 3. Which countries have 'land' in them? 
countries = ['Finland', 'Iceland', 'Ireland', 'Germany', 'Poland', 'Netherlands', 'Scotland']
land_countries = [c for c in countries if 'land' in c]
print(land_countries)
```

## Exercises: Level 2

```
# 1. Items unpacking
def unpack_country_info(country_info):
    country, capital, *rest = country_info
    return country, capital, rest

print(unpack_country_info(['Finland', 'Helsinki', 'Scandinavian', 5500000]))
```

## Exercises: Level 3

```
# 1. Write a function to handle ZeroDivisionError and ValueError
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return 'Cannot divide by zero'
    except TypeError:
        return 'Please provide numbers'

print(divide(10, 2))
print(divide(10, 0))
print(divide(10, 'a'))
```
