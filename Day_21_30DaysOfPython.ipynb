## Methods in Classes

```
class Person:
    def __init__(self, firstname, lastname, age, country, city):
        self.firstname = firstname
        self.lastname = lastname
        self.age = age
        self.country = country
        self.city = city

    def info(self):
        return f'{self.firstname} {self.lastname} is {self.age} years old.'

    def add_skill(self, skill):
        if not hasattr(self, 'skills'):
            self.skills = []
        self.skills.append(skill)

p1 = Person('Asabeneh', 'Yetayeh', 250, 'Finland', 'Helsinki')
p2 = Person('David', 'Smith', 25, 'UK', 'London')
print(p1.info())
print(p2.info())
p1.add_skill('Python')
p1.add_skill('JavaScript')
print(p1.skills)
```

## Inheritance

```
class Student(Person):
    def __init__(self, firstname, lastname, age, country, city, student_id):
        super().__init__(firstname, lastname, age, country, city)
        self.student_id = student_id

    def student_info(self):
        return f'{self.firstname} {self.lastname} ID: {self.student_id}'

s1 = Student('Eyob', 'Yetayeh', 30, 'Finland', 'Helsinki', 'A001')
s2 = Student('Lidiya', 'Teklemariam', 28, 'Finland', 'Espoo', 'A002')
print(s1.info())
print(s1.student_info())
print(s2.student_info())
```

## Exercises: Level 1

```
# 1. Create an Animal class
class Animal:
    def __init__(self, name, legs):
        self.name = name
        self.legs = legs

    def __str__(self):
        return f'{self.name} has {self.legs} legs'

a = Animal('Cat', 4)
print(a)
```

```
# 2. Subclass Dog and Cat from Animal
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, 4)

    def bark(self):
        return 'Woof!'

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, 4)

    def meow(self):
        return 'Meow!'

dog = Dog('Rover')
cat = Cat('Whiskers')
print(dog, dog.bark())
print(cat, cat.meow())
```

## Exercises: Level 2

```
# 1. Override the __str__ method
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'Name: {self.name}, Age: {self.age}'

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

p = Person('Asabeneh', 250)
print(p)
print(repr(p))
```

## Exercises: Level 3

```
# 1. Complete the PersonAccount class
class PersonAccount:
    def __init__(self, firstname, lastname, incomes, expenses):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = incomes
        self.expenses = expenses

    def total_income(self):
        return sum(self.incomes)

    def total_expense(self):
        return sum(self.expenses)

    def account_info(self):
        return (f'{self.firstname} {self.lastname}
'
                f'Total Income: {self.total_income()}
'
                f'Total Expense: {self.total_expense()}
'
                f'Account balance: {self.total_income() - self.total_expense()}')

account = PersonAccount('Asabeneh', 'Yetayeh', [1000, 2000, 3000], [500, 200])
print(account.account_info())
```
