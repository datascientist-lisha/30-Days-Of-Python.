## Arithmetic Operators

```
print(3 + 4)
print(3 - 4)
print(3 * 4)
print(3 / 4)
print(3 ** 4)
print(3 % 4)
print(3 // 4)
```

## Comparison Operators

```
print(3 > 2)
print(3 >= 3)
print(3 < 2)
print(2 < 3)
print(2 <= 3)
print(3 == 3)
print(3 != 2)
```

## Logical Operators

```
print(3 > 2 and 4 > 3)
print(3 > 2 and 3 < 2)
print(3 > 2 or 3 < 2)
print(not 3 > 2)
print(not True)
print(not False)
```

## Exercises: Level 2

```
# 1. Write a script that prompts the user to enter base and height
# and calculate the area of this triangle (area = 0.5 * b * h)
base = float(input('Enter base: '))
height = float(input('Enter height: '))
area = 0.5 * base * height
print('Area of the triangle:', area)
```

```
# 2. Write a script that prompts the user to enter side a, b, and c
# Calculate the perimeter of the triangle
a = float(input('Enter side a: '))
b = float(input('Enter side b: '))
c = float(input('Enter side c: '))
perimeter = a + b + c
print('Perimeter of the triangle:', perimeter)
```

```
# 3. Get length and width using prompt and calculate area and perimeter of rectangle
length = float(input('Enter length: '))
width = float(input('Enter width: '))
area = length * width
perimeter = 2 * (length + width)
print('Area of rectangle:', area)
print('Perimeter of rectangle:', perimeter)
```

```
# 4. Get radius using prompt and calculate the area and circumference of a circle
import math
radius = float(input('Enter radius: '))
area = math.pi * radius ** 2
circumference = 2 * math.pi * radius
print('Area of circle:', area)
print('Circumference of circle:', circumference)
```

```
# 5. Calculate slope, x-intercept and y-intercept of y = 2x - 2
# slope = 2, y-intercept = -2, x-intercept = 1
slope = 2
y_intercept = -2
x_intercept = -y_intercept / slope
print('Slope:', slope)
print('x-intercept:', x_intercept)
print('y-intercept:', y_intercept)
```

## Exercises: Level 3

```
# 1. Calculate the slope between two points (2,10) and (6,10)
x1, y1 = 2, 10
x2, y2 = 6, 10
slope = (y2 - y1) / (x2 - x1) if x2 != x1 else 'undefined'
print('Slope:', slope)
```

```
# 2. Compare the slopes in exercises 5 and 6
slope1 = 2
slope2 = (10 - 10) / (6 - 2)
print('Slope 1:', slope1)
print('Slope 2:', slope2)
print('Are they equal?', slope1 == slope2)
```
