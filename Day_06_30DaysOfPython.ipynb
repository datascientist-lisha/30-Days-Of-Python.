## Creating Tuples

```
empty_tuple = ()
print(type(empty_tuple))
fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits)
print(len(fruits))
```

## Accessing Tuple Items

```
fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits[0])
print(fruits[-1])
print(fruits[1:3])
```

## Tuple Methods

```
fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits.count('banana'))
print(fruits.index('orange'))
```

## Converting List to Tuple

```
fruits = ['banana', 'orange', 'mango', 'lemon']
fruits_tuple = tuple(fruits)
print(fruits_tuple)
print(type(fruits_tuple))
```

## Exercises: Level 2

```
# 1. Change the fruits tuple to a list, sort it and change it back to tuple
fruits = ('banana', 'orange', 'mango', 'lemon')
fruits_list = list(fruits)
fruits_list.sort()
fruits = tuple(fruits_list)
print(fruits)
```

```
# 2. Unpack the first three countries and the rest as scandic countries
countries = ('Germany', 'France', 'Belgium', 'Sweden', 'Denmark', 'Finland', 'Norway', 'Iceland')
ge, fr, be, *scandic = countries
print(ge, fr, be)
print(scandic)
```

## Exercises: Level 3

```
# 1. Create a tuple of tuples (nested tuples)
countries = (('Finland', 'Helsinki'), ('Sweden', 'Stockholm'), ('Norway', 'Oslo'))
print(countries)
for country, capital in countries:
    print(f'{country}-{capital}')
```

```
# 2. Change the countries tuple to a dictionary
countries_dict = {}
for country, capital in countries:
    countries_dict[country] = capital
print(countries_dict)
```
