## Creating Sets

```
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24]
print(it_companies)
print(A)
print(B)
```

## Set Methods

```
A = {19, 22, 24, 20, 25, 26}
A.add(30)
print(A)
A.update([31, 32, 33])
print(A)
A.remove(30)
print(A)
A.discard(100)
print(A)
```

## Set Operations

```
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
print(A.symmetric_difference(B))
print(A.issubset(B))
print(A.issuperset(B))
print(A.isdisjoint(B))
```
